```python
#데이터 프레임
#데이터 프레임은 2차원 배열이다. 행과 열로 만들어 지는 2차원 배열 구조는 
#MS 엑셀이나 데이터베이트 등 컴퓨터 관련 분야에서 다양하게 사용하고 있다

# 행 : 가로, Row, index
# 열 : 세로, Column, Data

#데이터 프레임 만들기
# 딕셔너리 -> 데이터프레임 변환
# 데이터프레임 = 판다스모듈.DataFrame(딕셔너리)

#pandas 모듈 불러오기
import pandas as pd

#딕셔너리 선언
dict_data = {
    "c0":[1,2,3],
    "c1":[4,5,6],
    "c2":[7,8,9],
    "c3":[10,11,12],
    "c3":[13,14,15],
}

#판다스 DataFrame() 함수로 딕셔너리를 데이터 프레임 형식으로 변환
df = pd.DataFrame(dict_data)
print("데이터프레임 원본")
print(df)
print("\n")

#데이터 프레임 자료형 출력
print(type(df))

#데이터프레임 객체 출력
print(df)
```

```python
#행 인덱스 / 열 이름 설정
#데이터 프레임 구조적 특성 상 2차원 배열 형태의 데이터를 데이터 프레임으로 변환하기 쉽다.
#2차원 배열에는 여러개 리스트(또는 투플)를 원소로 갖는 리스트(또는 투플)를 예로 들 수 있다

#한편 2차원 배열을 DataFrame()함수 인자로 전달하여 데이터 프레임으로 변환 할 때 
#행 인덱스와 열 이름 속성을 사용자가 직접 설정 할 수 있다.

#행 인덱스 이름 설정
# 판다스모듈.DataFrame(2차원배열, index=행인덱스 배열, columne=열이름 배열)

#pandas 모듈 불러오기
import pandas as pd

#행 인덱스 배열 / 열 이름 지정하여 데이터 프레임 만들기
df = pd.DataFrame(
    [
        [15,"남","덕영중"],
        [17,"여","수리중"]
    ],
    index=["준서","예은"],
    columns=["나이","성별","학교"]
)
print("데이터프레임 원본")
print(df)
print("\n")

#객체 출력
print(df)

# 행 인덱스
print(df.index)

# 열 이름
print(df.values)

#행 인덱스 / 열 이름 변경하기
#데이터 프레임 접근과 같은 방식으로 데이터 수정하는 방식으로 변경이 가능하다
df.index = ["학생1", "학생2"]
df.columns=["연령","남녀","소속"]
print(df.index)
print(df.values)
```

```python
# 행 인덱스 / 열 이름 변경
#rename() 메소드 사용하여 일부만 변경 할 수도 있다
#단 원본 객체를 직접 수정 하는게 아니라 새로운 객체에 임시로 변경 한 사항이라
#적용하려면 inplace=Ture 옵션을 적용 해줘야 한다

#pandas 모듈 불러오기
import pandas as pd

#행 인덱스 배열 / 열 이름 지정하여 데이터 프레임 만들기
df = pd.DataFrame(
    [
        [15,"남","덕영중"],
        [17,"여","수리중"]
    ],
    index=["준서","예은"],
    columns=["나이","성별","학교"]
)
print("데이터프레임 원본")
print(df)
print("\n")

#열 이름 중 나이를 연령으로, 성별을 남녀로, 학교를 소속으로 변경
df.rename(columns = {"나이":"연령","성별":"남녀","학교":"소속" }, inplace=True)

#행 인덱스 중에서 준서를 학생1로, 예은을 학생2로 변경
df.rename(index = {"준서":"학생1","예은":"학생2"}, inplace=True)
```

```python
#행 열의 위치 바꾸기
#데이터 프레임의 행과 열을 바꾸는 방법이다

#데이터 프레임 객체.transpose()
#데이터 프레임 객체.T

#판다스 모듈 추가
import pandas as pd

#데이터프레임 생성
dict_data = {
    '이름':["서준","우현","인아"],
    '수학':[90,80,70],
    '영어':[98,89,95],
    '음악':[85,95,100],
    '체육':[100,90,90]
}
df = pd.DataFrame(dict_data)
print("데이터프레임 원본")
print(df)
print("\n")

#데이터 프레임 전치하기
df=df.transpose()
print(df)

#데이터 프레임 다시 전치하기
df=df.T
print(df)
```

```python
#행 인덱스 재배열
#reindex() 메소드를 사용하면 데이터프레임의 행 인덱스를 새로운 배열로 재지정 할 수 있다

#새로운 배열로 행 인덱스를 재지정
#데이터 프레임 객체를 지정.reindex(새로운 인덱스 배열)

#판다스 모듈 추가
import pandas as pd

#데이터프레임 생성
dict_data = {
    "c0" : [1,2,3],
    "c1" : [4,5,6],
    "c2" : [7,8,9],
    "c3" : [10,11,12],
    "c4" : [13,14,15]
}

#딕셔너리를 데이터프레임으로 변환, 인덱스를 ["r0","r1","r2"]
df=pd.DataFrame(dict_data, index=["r0","r1","r2"])
print("딕셔너리를 데이터프레임으로 변환")
print(df)
print("\n")

#인덱스를 ["r0","r1","r2","r3","r4"]로 재지정
new_index = ["r0","r1","r2","r3","r4"]
ndf = df.reindex(new_index)
print("인덱스를 재지정")
print(ndf)
print("\n")

#새로운 인덱스 추가로 인하여 Null Data (NaN)를 채우기
new_index = ["r0","r1","r2","r3","r4"]
ndf = df.reindex(new_index, fill_value=0)
print("인덱스를 재지정 후 데이터 채우기")
print(ndf)
print("\n")
```

```python
#행 인덱스 초기화
#reset_index() 메소드를 활용하여 행 인덱스를 정수형 위치 인덱스로 초기화한다.

#정수 위치 인덱스로 초기화
#데이터 프레임 객체.reset_index()

#판다스 모듈 추가
import pandas as pd

#데이터프레임 생성
dict_data = {
    "c0" : [1,2,3],
    "c1" : [4,5,6],
    "c2" : [7,8,9],
    "c3" : [10,11,12],
    "c4" : [13,14,15]
}

#딕셔너리를 데이터프레임으로 변환, 인덱스를 ["r0","r1","r2"]
df=pd.DataFrame(dict_data, index=["r0","r1","r2"])
print("딕셔너리를 데이터프레임으로 변환")
print(df)
print("\n")

#행 인덱스를 정수형 인덱스로 초기화
df = df.reset_index()
print("행 인덱스를 정수형 인덱스로 초기화")
print(df)
print("\n")
```

```python
# 행 인덱스를 기준으로 데이터프레임 정렬
# sort_index() 메소드를 활용하여 행 인덱스를 기준으로 데이터 프레임의 값을 정렬한다
# ascending 옵션을 사용하여 오름/내림 차순으로 정렬
# ascending=True -> 오름차순
# ascending=False -> 내림차순

#행 인덱스를 기준으로 데이터 프레임 정렬
#데이터프레임 객체.sort_index()

#판다스 모듈 추가
import pandas as pd

#데이터프레임 생성
dict_data = {
    "c0" : [1,2,3],
    "c1" : [4,5,6],
    "c2" : [7,8,9],
    "c3" : [10,11,12],
    "c4" : [13,14,15]
}

#딕셔너리를 데이터프레임으로 변환, 인덱스를 ["r0","r1","r2"]
df=pd.DataFrame(dict_data, index=["r0","r1","r2"])
print("딕셔너리를 데이터프레임으로 변환")
print(df)
print("\n")

#행 인덱스를 내림차순으로 정렬
ndf = df.sort_index(ascending=False)
print("행 인덱스를 내림차순으로 정렬")
print(ndf)
print("\n")

#행 인덱스를 오름차순으로 정렬
ndf = df.sort_index(ascending=True)
print("행 인덱스를 오름차순으로 정렬")
print(ndf)
print("\n")
```

```python
# 열 데이터 값을 기준으로 데이터프레임 정렬
# 데이터프레임 객체,sort_values()
# ascending 옵션을 사용하여 오름/내림 차순으로 정렬
# ascending=True -> 오름차순
# ascending=False -> 내림차순

#판다스 모듈 추가
import pandas as pd

#데이터프레임 생성
dict_data = {
    "c0" : [1,2,3],
    "c1" : [4,5,6],
    "c2" : [7,8,9],
    "c3" : [10,11,12],
    "c4" : [13,14,15]
}

#딕셔너리를 데이터프레임으로 변환, 인덱스를 ["r0","r1","r2"]
df=pd.DataFrame(dict_data, index=["r0","r1","r2"])
print("딕셔너리를 데이터프레임으로 변환")
print(df)
print("\n")

#열 데이터를 내림차순으로 정렬
ndf = df.sort_values(by="c1",ascending=False)
print("열 데이터를 내림차순으로 정렬")
print(ndf)
print("\n")

#열 데이터를 오름차순으로 정렬
ndf = df.sort_values(by="c1",ascending=True)
print("열 데이터를 오름차순으로 정렬")
print(ndf)
print("\n")
```

```python
#시리즈연산
#시리즈 객체에 어떤 숫자를 더하면 시리즈 개별 원소에 각각 숫자를 더하고 계산한 결과를 시리즈 객체로 반환한다.
#사칙연산 가능
#시리즈 객체 + 연산자 + 숫자

#판다스 모듈 추가
import pandas as pd

#딕셔너리 데이터로 판다스 시리즈 만들기
series_data = pd.Series(
    {
        "국어" : 100,
        "영어" : 80,
        "수학" : 90
    }
)
print("딕셔너리 데이터로 판다스 시리즈 만들기")
print(series_data)
print("\n")

#학생의 과목별 점수를 200으로 나누기
div_score = series_data / 200

print("학생의 과목별 점수를 200으로 나누기")
print(div_score)
print("\n")
```

```python
#시리즈 끼리 연산
#시리즈와 시리즈 사이에 사칙연산을 처리하는 방법이다.

# 시리즈객체 + 연산자 + 시리즈객체

#판다스 모듈 추가
import pandas as pd

#딕셔너리 데이터로 판다스 시리즈 만들기
series_data_1 = pd.Series(
    {
        "국어" : 100,
        "영어" : 80,
        "수학" : 90
    }
)

series_data_2 = pd.Series(
    {
        "국어" : 80,
        "영어" : 90,
        "수학" : 80
    }
)
print("딕셔너리 데이터로 판다스 시리즈 만들기")
print("Series_1")
print(series_data_1)
print("\n")
print("Series_2")
print(series_data_2)
print("\n")

#두 시리그 객체의 과목별 점수로 사칙연산
add_res = series_data_1 + series_data_2
print("덧셈 결과")
print(add_res)
print("\n")

sud_res = series_data_1 - series_data_2
print("뺄셈 결과")
print(add_res)
print("\n")

mul_res = series_data_1 * series_data_2
print("곱셈 결과")
print(mul_res)
print("\n")

div_res = series_data_1 / series_data_2
print("나눗셈 결과")
print(div_res)
print("\n")
```

```python
# NaN값이 있는 시리즈 객체 연산
# 시리즈 끼리 연산 할 때 한쪽에 없는 인덱스나 인덱스가 있어도 NaN값인 상태로 연산 할때 판다스에서는 결과를 NaN으로 출력한다.

#판다스 모듈 추가
import pandas as pd

#numpy 모듈 추가
import numpy as np

#딕셔너리 데이터로 판다스 시리즈 만들기
series_data_1 = pd.Series(
    {
        "국어" : np.nan,
        "영어" : 80,
        "수학" : 90
    }
)

series_data_2 = pd.Series(
    {
        "수학" : 80,
        "국어" : 90
    }
)
print("딕셔너리 데이터로 판다스 시리즈 만들기")
print("Series_1")
print(series_data_1)
print("\n")
print("Series_2")
print(series_data_2)
print("\n")

#두 시리즈 객체의 과목별 점수 사칙연산 수행
add_res = series_data_1 + series_data_2
print("덧셈 결과")
print(add_res)
print("\n")

sud_res = series_data_1 - series_data_2
print("뺄셈 결과")
print(add_res)
print("\n")

mul_res = series_data_1 * series_data_2
print("곱셈 결과")
print(mul_res)
print("\n")

div_res = series_data_1 / series_data_2
print("나눗셈 결과")
print(div_res)
print("\n")
```

```python
#연산 메소드 활용
#단순 사칙연산 기호를 사용하게 되면 NaN 데이터가 있는 인덱스나 공용된 인덱스가 없는 부분은 NaN으로 반환된다.
#이때 연산 메소드를 사용하여 옵션에 fill_value를 사용하여 NaN 대신 채울 데이터를 지정 해 줄 수 있다

#연산 메소드 사용(시리즈와 시리즈 연산)
#시리즈 객체.연산메소드(시리즈 객체, fill_value=NaN 대신 채울 데이터)

#판다스 모듈 추가
import pandas as pd

#numpy 모듈 추가
import numpy as np

#딕셔너리 데이터로 판다스 시리즈 만들기
series_data_1 = pd.Series(
    {
        "국어" : np.nan,
        "영어" : 80,
        "수학" : 90
    }
)

series_data_2 = pd.Series(
    {
        "수학" : 80,
        "국어" : 90
    }
)
print("딕셔너리 데이터로 판다스 시리즈 만들기")
print("Series_1")
print(series_data_1)
print("\n")
print("Series_2")
print(series_data_2)
print("\n")

#두 시리즈 객체의 과목별 점수 연산 메소드 사용, NaN을 0으로 채우기
add_res = series_data_1.add(series_data_2, fill_value=0 )
print("덧셈 결과")
print(add_res)
print("\n")

sud_res = series_data_1.sub(series_data_2, fill_value=0 )
print("뺄셈 결과")
print(add_res)
print("\n")

mul_res = series_data_1.mul(series_data_2, fill_value=0 )
print("곱셈 결과")
print(mul_res)
print("\n")

div_res = series_data_1.div(series_data_2, fill_value=0 )
print("나눗셈 결과")
print(div_res)
print("\n")
```

```python
#HTML 웹 페이지에서 표 속성 가져오기
#판다스의 read_html()함수는 웹 페이지에 있는 <table>의 태그에서 표 형식의 데이터를 모두 찾아서 데이터 프레임으로 변환 한다.
# 표 데이터들은 각각 별도의 데이터프레임으로 변환되기 때문에 여러개의 데이터프레임으로 원소로 갖는 리스트가 반환된다.

#HTML 속성 표 읽기
#판다스모듈.read_html(웹주소 또는 html파일 로컬주소)

#판다스 모듈 추가
import pandas as pd

#HTML 파일의 경로 또는 웹페이지
url = "./data/sample.html"
tables = pd.read_html(url)

#표(table) 개수 확인
print(len(tables))

#table 리스트의 원소를 iteration 하면서 각가 화면에 출력
for count in range(len(tables)):
    print("tables[%s]" % count)
    print(tables[count])
    print("\n")

#파이썬 패키지 정보가 들어있는 두번째 데이터프레임을 선택하여 df변수에 저장
df = tables[1]

#name 열을 인덱스로 지정
df.set_index(["name"], inplace=True)
print(df)
```

```python
#웹 스크래핑
#BeautifylSoup들 웹 스크래핑 도구로 수집한 데이터를 판다스 데이터프레임 형식으로 정리하는 과정
#웹에서 크롤링->파이썬의 리스트나 딕셔너리->데이터프레임변환

#각종 모듈과 라이브러리 호출
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd

#위키피디아 마국 ETF웹 페이지에서 필요한 정보를 크롤링
#딕셔너리 형태로 변수 etfs에 저장
url = "https://en.wikipedia.org/wiki/List_of_American_exchange-traded_funds"
resp = requests.get(url)
soup = BeautifulSoup(resp.text, "lxml")
rows = soup.select("div > ul > li")

etfs = {}

for row in rows:
    try:
        efs_name = re.findall("^(.*) \(NYSE",row.text)
        efs_market = re.findall("\((.*)\|",row.text)
        efs_ticker = re.findall("NYSE Arca\|(.*)\)",row.text)
        
        if (len(efs_ticker) > 0 ) & (len(efs_market) > 0 ) & (len(efs_name) > 0 ):
            etfs[efs_ticker[0]] = [efs_market[0],efs_name[0]]
    
    except AttributeError as err:
        pass

#etfs 딕셔너리 출력
print(etfs)
print("\n")

#etfs 딕셔너리를 데이터프레임으로 변환
df = pd.DataFrame(etfs)
print(df)
```

```python
#파이차트

import pandas as pd
import matplotlib.pyplot as plt

plt.style.use("classic")

df=pd.read_csv("./data/auto-mpg.csv",header=None)

#열 이름 지정
df.columns = [
    "mpg",
    "cylinders",
    "displacement",
    "horsepower",
    "weight",
    "acceleration",
    "model_year",
    "origin",
    "name"
]

#데이터 개수 카운트 위해 값 1을 가진 열 추가
df["count"] = 1
df_origin = df.groupby("origin").sum()

#제조국가(origin) 값을 실제 지역명으로 변경
df_origin.index = ["USA","EU","JPN"]

#제조국가 열에 대한 파이차트 그리기
df_origin["count"].plot(
    kind="pie",
    figsize=(7,5),
    autopct="%1.1f%%",
    startangle=10,
    colors=["chocolate","bisque","cadetblue"]
)

plt.title("Model Origin",size=20)
plt.axis("equal")
plt.legend(labels=df_origin.index, loc="upper right")
plt.show()
```

```python
#박스플롯

import pandas as pd
import matplotlib.pyplot as plt

plt.style.use("classic")

df=pd.read_csv("./data/auto-mpg.csv",header=None)

#열 이름 지정
df.columns = [
    "mpg",
    "cylinders",
    "displacement",
    "horsepower",
    "weight",
    "acceleration",
    "model_year",
    "origin",
    "name"
]

#그래프 객체 생성, 
fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

#axe객체에 boxplot 메소드로 그래프 출력
ax1.boxplot(
    x=[
        df[df["origin"]==1]["mpg"],
        df[df["origin"]==2]["mpg"],
        df[df["origin"]==3]["mpg"]
    ],
    labels=["USA","EU","JPN"]
)

ax2.boxplot(
    x=[
        df[df["origin"]==1]["mpg"],
        df[df["origin"]==2]["mpg"],
        df[df["origin"]==3]["mpg"]
    ],
    labels=["USA","EU","JPN"],
    vert=False
)

ax1.set_title("Fuel economy distribution by made country(vertical boxplot)")
ax2.set_title("Fuel economy distribution by made country(horizontal boxplot)")

plt.show()
```

```python
#지도에 마커 표기

#모듈 등록
import folium
import pandas as pd

#서울지역 대학교 정보 불러오기
df = pd.read_excel("./data/서울지역 대학교 위치.xlsx",engine="openpyxl")

#서울지도 생성
seoul_map = folium.Map(location=[37.55,126.98], tiles="stamen Terrain",zoom_start=12)

#대학교 위치 마커로 표기
for name,lat,lng in zip(df.index, df.위도, df.경도):
    folium.Marker([lat,lng], popup=name).add_to(seoul_map)

#지도를 HTML file로 저장하기
seoul_map.save("./data/seoul_map_3.html")
```

```python
#지도에 원형마커 표기

#모듈 등록
import folium
import pandas as pd

#서울지역 대학교 정보 불러오기
df = pd.read_excel("./data/서울지역 대학교 위치.xlsx",engine="openpyxl")

#서울지도 생성
seoul_map = folium.Map(location=[37.55,126.98], tiles="stamen Terrain",zoom_start=12)

#대학교 위치 마커로 표기
for name,lat,lng in zip(df.index, df.위도, df.경도):
    folium.Marker(
        [lat,lng],
        radius=10,
        color="brown",
        fill=True,
        fill_color="coral",
        fiil_opacity=0.7,
        popup=name
    ).add_to(seoul_map)

#지도를 HTML file로 저장하기
seoul_map.save("./data/seoul_map_4.html")
```

```python
#지도 영역에 단계 구분도 표시

#모듈 등록
import folium
import pandas as pd
import json

file_path = "./data/경기도인구데이터.xlsx"
df = pd.read_excel(file_path, index_col="구분", engine="openpyxl")
df.columns = df.columns.map(str)

geo_path = "./data/경기도행정구역경계.json"
try:
    geo_data = json.load(open(geo_path,encoding="utf-8"))
except:
    geo_data = json.load(open(geo_path,encoding="utf-8-sig"))

#경기도 지도 만들기
g_map = folium.Map(location=[37.5502,126.982], tiles="stamen Terrain",zoom_start=9)
    
#출력할 연도 선택
year = "2007"

#choropleth 클래스로 단계구분 표시하기
folium.Choropleth(
    geo_data = geo_data,
    data = df[year],
    columns = [df.index , df[year]],
    fill_color="YlOrRd",fiil_opacity=0.7,line_opacity=0.3,
    threshold_scale=[10000,100000,300000,500000,700000],
    key_on="feature.properties.name"
).add_to(g_map)

#지도를 HTML파일로 저장하기
g_map.save("./data/gyonggi_poulation_"+year+".html")
```
